```python
from matplotlib import pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv("Default_FiniteKey.csv")
```

```python
df1 = pd.read_csv("FiniteKey_1dB.csv")
df2 = pd.read_csv("FiniteKey_2dB.csv")
```

```python
df2
```

```python
plt.plot(df["max_elev"],df["FiniteKey"], label="dfault simulation")
plt.plot(df1["max_elev"],df1["FiniteKey"],label="additional 1dB loss")
plt.plot(df2["max_elev"],df2["FiniteKey"],label="additional 2dB loss")
plt.xlabel("Maximum Elevation",size=16)
plt.ylabel("Secret Key size",size=16)
plt.title("Finite key for passes with different maximum elevations",size=16)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.legend()
plt.show()
```
