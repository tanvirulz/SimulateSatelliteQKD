```python
from matplotlib import pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv("Default_FiniteKey.csv")
```

```python
df1 = pd.read_csv("FiniteKey_1dB.csv")
df2 = pd.read_csv("FiniteKey_2dB.csv")
```

```python
df
```

```python
plt.figure(figsize=(7,5))
plt.plot(df["max_elev"],df["FiniteKey"]/1000, label="dfault simulation")
plt.plot(df1["max_elev"],df1["FiniteKey"]/1000,label="additional 1dB loss")
plt.plot(df2["max_elev"],df2["FiniteKey"]/1000,label="additional 2dB loss")
plt.xlabel("Maximum Elevation",size=16)
plt.ylabel("Secret Key size (kilo bits)",size=16)
plt.title("Finite key for passes with different maximum elevations",size=16)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.legend(prop={'size': 14})
plt.savefig("FiniteKey.png")
plt.show()
```
